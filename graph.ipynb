```python
import neo4jupyter
from py2neo import Graph, Node, Relationship,NodeMatcher
neo4jupyter.init_notebook_mode()
graph = Graph("bolt://localhost:7687",auth=("neo4j","86360365jkl"))
graph.delete_all()
```

```python
#遍历节点
for i in range(1,len(df)):
    matcher = NodeMatcher(graph)
    nodelist = list(matcher.match('案件名称', name=df['案件名称'].tolist()[i]))
    if len(nodelist) > 0:
        matcher = NodeMatcher(graph)
        nodelist1 = list(matcher.match(df['实体属性'].tolist()[i], name=df['实体'].tolist()[i]))
        if len(nodelist1)>0:
            zhucong = Relationship(nodelist[0],df['实体属性'].tolist()[i], nodelist1[0])
            graph.create(zhucong)

        else:
            cong = Node(df['实体属性'].tolist()[i], name=df['实体'].tolist()[i])
            graph.create(cong)
            zhucong = Relationship(nodelist[0], df['实体属性'].tolist()[i], cong)
            graph.create(zhucong)
    else:
        zhu = Node('案件名称', name=df['案件名称'].tolist()[i])
        graph.create(zhu)
        matcher = NodeMatcher(graph)
        nodelist1 = list(matcher.match(df['实体属性'].tolist()[i], name=df['实体'].tolist()[i]))
        if len(nodelist1)>0:
            zhucong = Relationship(zhu, df['实体属性'].tolist()[i], nodelist1[0])
            graph.create(zhucong)
        else:
            cong = Node(df['实体属性'].tolist()[i], name=df['实体'].tolist()[i])
            graph.create(cong)
            zhucong = Relationship(zhu, df['实体属性'].tolist()[i], cong)
            graph.create(zhucong)
```

```python
neo4jupyter.draw(graph, options = {"案件名称": "name","审理法院":"name"})
```
